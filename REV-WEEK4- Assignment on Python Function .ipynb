Problem 4:
Write a function that will take in an unknown number of arguments and multiply all of them together and run the function for these 2 sets of numbers:
1,2,3,4,5
12,13,14    

Sample Output:
120

```

### CODE4:

````python
def multiply_all(*x):
    s = 1
    for i in x:
        s = s*i 
    return s

print(multiply_all(1,2,3,4,5))
print(multiply_all(12,13,14))
````

### CODE6:

````python
dict = {"First Name":"John", "Last Name":"Wick","Nickname":"Jojo", "Middle Name":"Candle"}

def all_name(**kwargs):
    for i,val in kwargs.items():
        print(f'{i} is {val}')
        

print(all_name(**dict))
````

Write a recursive function that will sum all numbers from 1 to n. n is the argument of the function.


Sample Input:
4
Sample Output:
10

### CODE8:

````python
def summation(n):
    if n == 1:
        return n 
    else :
        return n+summation(n-1)
   
    
summation(4)
````
